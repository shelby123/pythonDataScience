# Multiple linear regression
### Can we predict the calorie content of a cereal? 

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("cereals.csv")
```

```python
y = df.calories.values
X = df[['sugars', 'carbo', 'protein', 'fat', 'sodium', 'fiber', 'potass']].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.25, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
results = pd.DataFrame(y_test, columns= ['test'])
results['predicted'] = y_pred
```

```python
results
```

# Can we do it with SVR? 

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler()
X_scaled = X_scaler.fit_transform(X_train)
X_test_scaled = X_scaler.transform(X_test)

y_scaler = StandardScaler()
y_scaled = y_scaler.fit_transform(y_train.reshape((-1, 1)))
y_test_scaled = y_scaler.transform(y_test.reshape((-1, 1)))

rbf_regressor = SVR(kernel="rbf")
rbf_regressor.fit(X_scaled, y_scaled.reshape(-1))

linear_regressor = SVR(kernel = "linear")
linear_regressor.fit(X_scaled, y_scaled.reshape(-1))
```

```python
y_pred_rbf = rbf_regressor.predict(X_test_scaled)
y_pred_linear = linear_regressor.predict(X_test_scaled)
```

```python
y_pred_unscaled_rbf = y_scaler.inverse_transform(y_pred_rbf)
y_pred_unscaled_linear = y_scaler.inverse_transform(y_pred_linear)
```

```python
svr_results = pd.DataFrame(y_test, columns= ['test'])
svr_results['predicted RBF'] = y_pred_unscaled_rbf
svr_results['predicted linear'] = y_pred_unscaled_linear
svr_results
```

# What about Non linear regression?

```python
from sklearn.tree import DecisionTreeRegressor

tree_regressor = DecisionTreeRegressor()
tree_regressor.fit(X_train, y_train)
```

```python
y_pred_decisionTree = tree_regressor.predict(X_test)
```

```python
decisionTree_results = pd.DataFrame(y_test, columns= ['test'])
decisionTree_results['predicted decisionTree'] = y_pred_decisionTree
decisionTree_results
```

```python
from sklearn.ensemble import RandomForestRegressor

forest_regressor = RandomForestRegressor(n_estimators = 10, random_state=0)
forest_regressor.fit(X_train, y_train)

forest_regressor100 = RandomForestRegressor(n_estimators = 100, random_state=0)
forest_regressor100.fit(X_train, y_train)

forest_regressor500 = RandomForestRegressor(n_estimators = 500, random_state=0)
forest_regressor500.fit(X_train, y_train)
```

```python
y_pred_forest = forest_regressor.predict(X_test)
y_pred_forest100 = forest_regressor100.predict(X_test)
y_pred_forest500 = forest_regressor500.predict(X_test)
```

```python
forest_results = pd.DataFrame(y_test, columns= ['test'])
forest_results['predicted forest'] = y_pred_forest
forest_results['predicted forest 100'] = y_pred_forest100
forest_results['predicted forest 500'] = y_pred_forest500
forest_results
```

# Selecting the best model

- $R^2 = 1- \frac{SS_{res}}{SS_{avg}}$, Where $ SS_{res}=\sum(y_{pred} - y_{true})^2$, and $ SS_{avg}=\sum(y_{mean} - y_{true})^2$

```python
import random
def f(x):
    return (3*x+random.uniform(-5, 5))

def f1(x):
    return (3*x+random.uniform(-1, 1))

def f2(x):
    return (3*x+random.uniform(-10, 10))

def f3(x):
    return (x+random.uniform(-5, 5))
```

```python
from sklearn.linear_model import LinearRegression
functions = [f, f1, f2, f3]
for func in functions:
    X = np.arange(start = 0 , stop = 10, step=.1)
    y = list(map(func,X))
    regressorTest = LinearRegression()
    regressorTest.fit(X.reshape((-1, 1)), y)
    plt.scatter(X, y, color="blue")
    plt.plot(X, regressorTest.predict(X.reshape((-1, 1))), color="red")
    plt.title(str(regressorTest.score(X.reshape(-1, 1), y)))
    plt.show();
```
