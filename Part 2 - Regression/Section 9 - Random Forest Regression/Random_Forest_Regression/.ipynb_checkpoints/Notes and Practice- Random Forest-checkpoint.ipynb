```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("Position_Salaries.csv")
```

**Notes:**
- random forest is just an aggregation of decision trees
- random forest is a type of __ensemble__ learning

```python
X = df.Level.values.reshape((-1, 1))
y = df.Salary.values
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X, y)

regressor50 = RandomForestRegressor(n_estimators=50, random_state=0)
regressor50.fit(X, y)

regressor100 = RandomForestRegressor(n_estimators=100, random_state=0)
regressor100.fit(X, y)

regressor200 = RandomForestRegressor(n_estimators=200, random_state=0)
regressor200.fit(X, y)

regressor500 = RandomForestRegressor(n_estimators=500, random_state=0)
regressor500.fit(X, y)
```

```python
y_preds = [regressor.predict([[6.5]]), 
           regressor50.predict([[6.5]]), 
           regressor100.predict([[6.5]]), 
           regressor200.predict([[6.5]]), 
           regressor500.predict([[6.5]])]
```

```python
y_preds
```

```python
X_grid = np.arange(start=min(X), stop=max(X), step=.01)
plt.figure(figsize=(20,10))
plt.scatter(X, y, color='black')
plt.plot(X_grid, regressor.predict(X_grid.reshape((-1,1))), 'red')
plt.plot(X_grid, regressor50.predict(X_grid.reshape((-1,1))), 'orange')
plt.plot(X_grid, regressor100.predict(X_grid.reshape((-1,1))), 'yellow')
plt.plot(X_grid, regressor200.predict(X_grid.reshape((-1,1))), 'green')
plt.plot(X_grid, regressor500.predict(X_grid.reshape((-1,1))), 'blue')
```

- note the convergence that begins to happen. Increasing the number of trees does NOT result in more steps in the graph. it just increases the number of trees being sampled, and so we converge on the average of the results
